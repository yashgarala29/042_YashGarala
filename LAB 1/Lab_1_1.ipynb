```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv('./Data_for_Transformation.csv')
plt.scatter(data['Age'],data['Salary'],marker='+')
plt.show()
```

```python
plt.hist(data['Salary'])
plt.show()
```

```python
df=pd.DataFrame(data,columns=['Country'])

val=df['Country'].value_counts()

labels=val.keys()
# print(val)
values = val
fig=plt.figure()
plt.pie(values,labels=labels,autopct='%.2f%%',startangle=90)
plt.axis('equal')
plt.show()
```
