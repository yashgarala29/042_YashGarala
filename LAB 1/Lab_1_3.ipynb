```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 1. Create Two numpy array of size 3 X 2 and 2 X 3
 2.  Randomly Initalize that array


```python
numpy_arr1=np.random.randint(100,size=(3,2))
numpy_arr2=np.random.randint(100,size=(2,3))
print(numpy_arr1)
print(numpy_arr2)
```

# Perform matrix multiplication

```python
multiplication= np.matmul(numpy_arr1,numpy_arr2)
print(multiplication)
```

# Perform elementwise matrix multiplication

```python
ans=np.zeros((len(numpy_arr1),len(numpy_arr2[0])))
# print(ans)
for i in range(len(numpy_arr1)):
   for j in range(len(numpy_arr2[0])):
       for k in range(len(numpy_arr2)):
           ans[i][j] += numpy_arr1[i][k] * numpy_arr2[k][j]

print(ans)
```

 Find mean of first matrix


```python
matrix_mean=np.mean(numpy_arr1)
print(matrix_mean)
```

# Convert Numeric entries(columns) of mtcars.csv to Mean Centered Version

```python
data=pd.read_csv('./mtcars.csv')
column_name=['mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
for i in column_name:
  print("-----------------",i,"------------------")
  arr1=np.array(data[i])
  npcentered=arr1-np.mean(arr1)
  print(npcentered)
```
